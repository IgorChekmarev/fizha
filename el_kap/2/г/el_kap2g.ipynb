```python
import pandas as pd

# Читаем Excel файл
df = pd.read_excel('2gg.xlsx')

# В данных первый столбец содержит "dp" и значения, нам нужны все столбцы
# Разделяем строки по пробелам и создаем новый DataFrame
data = []
for index, row in df.iterrows():
    # Получаем строку из первого (и единственного) столбца
    text = str(row[0])
    
    # Разбиваем по пробелам
    parts = text.strip().split()
    
    # Проверяем, что строка начинается с "dp" и содержит достаточно частей
    if len(parts) >= 4 and parts[0].lower() == 'dp':
        # Берем следующие 3 части (время, напряжение, ток)
        try:
            time = float(parts[1])
            voltage = float(parts[2])
            current = float(parts[3])
            data.append([time, voltage, current])
        except ValueError:
            print(f"Ошибка преобразования в строке: {text}")

# Создаем новый DataFrame
result_df = pd.DataFrame(data, columns=['Время, с', 'Напряжение, В', 'Ток, А'])
print(result_df)
# Сохраняем результат
result_df.to_csv('2ggready.csv', index=False, sep=';', encoding='utf-8-sig')
result_df.to_excel('2ggready.xlsx', index=False)

print(f"Извлечено строк: {len(result_df)}")
print("\nПервые 5 строк:")
print(result_df.head())
print("\nПоследние 5 строк:")
print(result_df.tail())
print("\nСтатистика:")
print(result_df.describe())
```
